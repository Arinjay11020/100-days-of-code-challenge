```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
def takeInput():
    inputList=[int(i) for i in input().split()]
    head=None
    for currData in inputList:
        if currData==-1:
            break
        newNode=Node(currData)
        if head is None:
            head=newNode
        else:
            curr=head
            while curr.next is not None:
                curr=curr.next
            curr.next=newNode
    return head
def takeInputBetter():
    inputList=[int(i) for i in input().split()]
    head=None
    tail=None
    for currData in inputList:
        if currData==-1:
            break
        newNode=Node(currData)
        if head is None:
            head=newNode
            tail=newNode
        else:
            tail.next=newNode
            tail=newNode
    return head
def printLL(head):
    while head is not None:
        print(str(head.data)+"->",end="")
        head=head.next
    print("None")
    return
def length(head):
    count=0
    while head is not None:
        count+=1
        head=head.next
    return count
def inode(head,i):
    if i<0 or i>=length(head):
        print("Invalid input")
    count=0
    curr=head
    while count<i:
        curr=curr.next
        count+=1
    print(curr.data)
    return
def insertAti(head,i,data):
    if i<0 or i>length(head):
        return head
    count=0
    prev=None
    curr=head
    newNode=Node(data)
    while count<i:
        prev=curr
        curr=curr.next
        count+=1
    if prev is not None:
        prev.next=newNode
        newNode.next=curr
    else:
        head=newNode
        head.next=curr
    return head
head=takeInput()
#head2=takeInputBetter()
printLL(head)
#printLL(head2)
print(length(head))
inode(head,3)
# insertAti(head,3,49)
# printLL(head)
printLL(insertAti(head,0,0))
printLL(head)
# insertAti(head,11,36)
# printLL(head)
```
